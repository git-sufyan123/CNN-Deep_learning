```python
import tensorflow as tf
from tensorflow.keras import layers,models
from tensorflow.keras.datasets import mnist

(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
#normalization and dimension correction
x_train = x_train / 255.0
x_test = x_test / 255.0

x_train = x_train[...,tf.newaxis]
x_test = x_test[...,tf.newaxis]

print(x_train.shape)
```

```python
#model
basic_model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
])
    
```

```python
basic_model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
basic_hist = basic_model.fit(x_train,y_train,validation_split = 0.2,epochs=5,batch_size=32)
```

```python
#batch_normalization + Dropout + augmentation

augmentation = models.Sequential([
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])

complex_model = models.Sequential([

    augmentation,
    layers.Conv2D(32,(3,3),input_shape=(28,28,1)),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.Dense(10),
    layers.Dropout(.5),
    layers.Softmax()
])
```

```python
#compilation

complex_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

complex_hist = complex_model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=15,
    batch_size=32
)
```

```python
print("Basic CNN")
print(basic_model.evaluate(x_test, y_test))

print("complex CNN")
print(complex_model.evaluate(x_test, y_test))
```

accuracy increased in our complex model and also loss is less than the basic model suggesting its more confident in its prediction
